###### The purpose of this project is to generate a signal for the growth of SPY, an exchange-traded fund which tracks the S&P 500 index. The signal is generated by applying a multiple linear regression model using multiple indices from common global markets. 

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# stock market data from SPY and 8 common global markets are imported into a Pandas DataFrame
aord = pd.read_csv('data/ALLOrdinary.csv', index_col=0)
nikkei = pd.read_csv('data/Nikkei225.csv', index_col=0)
hsi = pd.read_csv('data/HSI.csv', index_col=0)
daxi = pd.read_csv('data/DAXI.csv', index_col=0)
cac40 = pd.read_csv('data/CAC40.csv', index_col=0)
sp500 = pd.read_csv('data/SP500.csv', index_col=0)
dji = pd.read_csv('data/DJI.csv', index_col=0)
nasdaq = pd.read_csv('data/nasdaq_composite.csv', index_col=0)
spy = pd.read_csv('data/SPY.csv', index_col=0)
```

## Step 1: Data Munging

###### This project aims to predict the price change of SPY when the US market opens in the morning. Therefore, due to timezone issues, the existing stock market data has to be amended.
###### For US and European markets, the predictors are the opening price - previous day's opening price
###### For Asian markets, the predictors are the closing price - opening price (same day)
###### The response variable (SPY) is the next day's opening price - current opening price

```python
indicepanel=pd.DataFrame(index=spy.index)

indicepanel['spy']=spy['Open'].shift(-1)-spy['Open']
indicepanel['spy_lag1']=indicepanel['spy'].shift(1)
indicepanel['sp500']=sp500["Open"]-sp500['Open'].shift(1)
indicepanel['nasdaq']=nasdaq['Open']-nasdaq['Open'].shift(1)
indicepanel['dji']=dji['Open']-dji['Open'].shift(1)

indicepanel['cac40']=cac40['Open']-cac40['Open'].shift(1)
indicepanel['daxi']=daxi['Open']-daxi['Open'].shift(1)

indicepanel['aord']=aord['Close']-aord['Open']
indicepanel['hsi']=hsi['Close']-hsi['Open']
indicepanel['nikkei']=nikkei['Close']-nikkei['Open']
indicepanel['Price']=spy['Open']
```

```python
indicepanel.head()
```

###### As each market has a varying number of holidays,  it is expected that the combined DataFrame has significant numbers of null values. These missing values are handled by forward filling the data. The exeption is the first days data, which is dropped entirely since forward filling is not available in this case.

```python
indicepanel.isnull().sum()
```

```python
indicepanel = indicepanel.fillna(method='ffill')
indicepanel = indicepanel.dropna()
```

```python
indicepanel.isnull().sum()
```

```python
# Saving the data for future use
indicepanel.to_csv('data/indicepanel.csv')
```

## Step 2: Data Spliting

###### To improve model performance, the data is split into a training set and a test set, each consisiting of 1000 days of data. Data splitting will also help to prevent the model from becoming overfitted.

```python
Train = indicepanel.iloc[-2000:-1000, :]
Test = indicepanel.iloc[-1000:, :]
print(Train.shape, Test.shape)
```

## Step 3: Exploring the Training Data Set

###### Generating a scatter matrix allows for the observation of all assocations. From this viewpoint, it seems that there is very little correlation between SPY and other common markets.

```python
# Generate scatter matrix among all stock markets (and the price of SPY) to observe the association

sm = pd.plotting.scatter_matrix(Train, figsize=(10, 10))
```

## Step 4: Checking the Correlation between SPY and each other index

###### As shown above, it is seen that there is little correlation between SPY and any other index, aside from the All Ordinaries index (aord), where some correlation exists.

```python
corr_array = Train.iloc[:, :-1].corr()['spy']
print(corr_array)
```

```python
# The training data set is used to generate a linear model, using Ordinary Least Squares method. This method is provided in the
# statsmodels API.

formula = 'spy~spy_lag1+sp500+nasdaq+dji+cac40+aord+daxi+nikkei+hsi'
lm = smf.ols(formula=formula, data=Train).fit()
lm.summary()
```

###### Key notes from the summary:
###### F-statistic (model significance): with a value < 5%, this indicates that the model includes useful indicators
###### P-values: it is confirmed that - apart from aord - most of the predictors are not significant

## Step 5: Making a prediction

###### Predictions are made on both the training and test datasets. Whilst it is expected that the model will perfom better on the test set, predictions are made on both sets to ensure that there is no overfitting of the data.

###### Once predictions have been made, a scatter plot between the real daily change and predicated daily change of SPY is generated. As seen in the plot below, there is some positive correlation, which is a somewhat successful result.

```python
Train['PredictedY'] = lm.predict(Train)
Test['PredictedY'] = lm.predict(Test)
```

```python
plt.scatter(Train['spy'], Train['PredictedY'])
```

## Step 6: Model Evaluation - RMSE and Adjusted $R^{2}$

###### The performance of the model is measured using RMSE (prediction error) and Adjusted $R^{2}$ (explanation of variance).


```python
# RMSE - Root Mean Squared Error, Adjusted R^2
def adjustedMetric(data, model, model_k, yname):
    data['yhat'] = model.predict(data)
    SST = ((data[yname] - data[yname].mean())**2).sum()
    SSR = ((data['yhat'] - data[yname].mean())**2).sum()
    SSE = ((data[yname] - data['yhat'])**2).sum()
    r2 = SSR/SST
    adjustR2 = 1 - (1-r2)*(data.shape[0] - 1)/(data.shape[0] -model_k -1)
    RMSE = (SSE/(data.shape[0] -model_k -1))**0.5
    return adjustR2, RMSE
```

```python
def assessTable(test, train, model, model_k, yname):
    r2test, RMSEtest = adjustedMetric(test, model, model_k, yname)
    r2train, RMSEtrain = adjustedMetric(train, model, model_k, yname)
    assessment = pd.DataFrame(index=['R2', 'RMSE'], columns=['Train', 'Test'])
    assessment['Train'] = [r2train, RMSEtrain]
    assessment['Test'] = [r2test, RMSEtest]
    return assessment
```

```python
# Get the assement table fo our model
assessTable(Test, Train, lm, 9, 'spy')
```

###### From the results above, it can be concluded that the model is not overfitted, as there is no dramatic difference between statstics for the training and test sets.

# Step 7 - Strategy Evaluation

###### The follwing code cell caluculates the potential profits (vs a typical "buy and hold" strategy) of the signal-based strategy for both the training and test data sets. As shown below, the test set performs better as expected, however the training set is not too far behind, which is a positive result.

```python
# Train
Train['Order'] = [1 if sig>0 else -1 for sig in Train['PredictedY']]
Train['Profit'] = Train['spy'] * Train['Order']

Train['Wealth'] = Train['Profit'].cumsum()
print('Total profit made in Train: ', Train['Profit'].sum())

Test['Order'] = [1 if sig>0 else -1 for sig in Test['PredictedY']]
Test['Profit'] = Test['spy'] * Test['Order']

Test['Wealth'] = Test['Profit'].cumsum()
print('Total profit made in Test: ', Test['Profit'].sum())

plt.figure(figsize=(5, 5))
plt.title('Performance of Strategy in Train')
plt.plot(Train['Wealth'].values, color='green', label='Signal based strategy')
plt.plot(Train['spy'].cumsum().values, color='red', label='Buy and Hold strategy')
plt.legend()

plt.figure(figsize=(5, 5))
plt.title('Performance of Strategy in Test')
plt.plot(Test['Wealth'].values, color='green', label='Signal based strategy')
plt.plot(Test['spy'].cumsum().values, color='red', label='Buy and Hold strategy')
plt.legend()

plt.show()
```

### Practical Standards - Sharpe Ratio and Maximum Drawdown

###### Finally, the performance of the model is evaluated using two common practical standards: Sharpe ratio (risk-adjusted return) and Maximum drawdown (largest peak-trough drop)

```python
# Calculating overall wealth e.g. profit + intial investment

Train['Wealth'] = Train['Wealth'] + Train.loc[Train.index[0], 'Price']
Test['Wealth'] = Test['Wealth'] + Test.loc[Test.index[0], 'Price']

# Sharpe Ratio on Train data
Train['Return'] = np.log(Train['Wealth']) - np.log(Train['Wealth'].shift(1))
dailyr = Train['Return'].dropna()
print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))

# Sharpe Ratio on Test data
Test['Return'] = np.log(Test['Wealth']) - np.log(Test['Wealth'].shift(1))
dailyr = Test['Return'].dropna()
print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))

# Maximum Drawdown in Train data
Train['Peak'] = Train['Wealth'].cummax()
Train['Drawdown'] = (Train['Peak'] - Train['Wealth'])/Train['Peak']
print('Maximum Drawdown in Train is ', Train['Drawdown'].max())

# Maximum Drawdown in Test data
Test['Peak'] = Test['Wealth'].cummax()
Test['Drawdown'] = (Test['Peak'] - Test['Wealth'])/Test['Peak']
print('Maximum Drawdown in Test is ', Test['Drawdown'].max())
```

###### With a yearly Sharpe ratio > 2 (classified as adequate/good), the strategy can be considered a success. Similarly, a maximum drawdown < 20% is an acceptable value, and would ensure that the investor can trade with peace of mind.
